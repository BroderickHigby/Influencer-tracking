```python
cities = ca.City
```

```python
cities_counts = cities.value_counts().to_frame()
```

```python
cities_counts = cities_counts.reset_index()
cities_counts = cities_counts.rename(index=str, columns={'index':'city', 'City':'counts'})
cities_counts.head()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib.collections import LineCollection
from matplotlib.patches import Polygon
from matplotlib.patches import Polygon
from mpl_toolkits.basemap import Basemap

%matplotlib notebook

import shapefile
```

```python
def draw_cali():
    """
    Draws a map focused on California
    """
    
    fig = plt.figure()
    
    projection="merc"
    llcrnrlat=32
    urcrnrlat=42.5
    llcrnrlon=-126
    urcrnrlon=-114
    resolution='i'
    
    m = Basemap(projection=projection, llcrnrlat=llcrnrlat, urcrnrlat=urcrnrlat, llcrnrlon=llcrnrlon, urcrnrlon = urcrnrlon, resolution=resolution)
    
    m.drawcoastlines()
    m.drawmapboundary(fill_color='aqua',zorder=-11)
    m.fillcontinents(color='coral',lake_color='aqua',zorder=-10)
    m.drawcountries()
    m.drawcounties()
    m.drawstates()
    
    return m
```

```python
fig = plt.figure()

projection="merc"
llcrnrlat=32
urcrnrlat=42.5
llcrnrlon=-126
urcrnrlon=-114
resolution='i'

m = Basemap(projection=projection, llcrnrlat=llcrnrlat, urcrnrlat=urcrnrlat, llcrnrlon=llcrnrlon, urcrnrlon = urcrnrlon, resolution=resolution)

m.drawcoastlines()
m.drawmapboundary(fill_color='aqua',zorder=-11)
m.fillcontinents(color='coral',lake_color='aqua',zorder=-10)
m.drawcountries()
m.drawcounties()
m.drawstates()
```

```python
m = draw_cali()
plt.show()
```

```python
ca_cities = pd.read_csv('https://raw.githubusercontent.com/jakevdp/PythonDataScienceHandbook/master/notebooks/data/california_cities.csv')
```

```python
ca_cities.head()
```

```python
ca_locs = ca_cities[['city', 'latd', 'longd']]
ca_locs.head()
ca_locs.loc[ca_locs.city == 'Irvine', :]
```

```python
cities_full = cities_counts.join(ca_locs, lsuffix="0_")
```

```python
cities_full.head()
```

```python
cities_counts.head()
```

```python
cities_counts['city'] = cities_counts.city.replace(' ', '')
```

```python
cities_counts.head()
```

```python
cities_full = cities_counts.merge(ca_locs, left_on='city', right_on='city')
```

```python
cities_full.head()
cities_full.isnull().sum().sum()
```

```python
lat = cities_full.latd.values
lon = cities_full.longd.values
counts = cities_full.counts.values
```

```python
SCALE = 2

fig = plt.figure()

projection="merc"
llcrnrlat=32
urcrnrlat=42.5
llcrnrlon=-126
urcrnrlon=-114
resolution='i'

m = Basemap(projection=projection, llcrnrlat=llcrnrlat, urcrnrlat=urcrnrlat, llcrnrlon=llcrnrlon, urcrnrlon = urcrnrlon, resolution=resolution)

m.drawcoastlines()
# m.drawmapboundary(fill_color='aqua',zorder=-11)
# m.fillcontinents(color='coral',lake_color='aqua',zorder=-10)
m.drawcountries()
m.drawcounties()
m.drawstates()

m.scatter(lon, lat, latlon=True, s=counts*SCALE, alpha=0.5)
```
